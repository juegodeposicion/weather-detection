```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("D:/fy_project1/data1.csv")
print("Stored")
```

```python
#check for missing values
print("Count of null values per column:")
print(df.isnull().sum())
```

```python
#data types of each column
print("\nData types of each column:")
print(df.dtypes)
```

```python
print("\nDescriptive Statistics:")
print(df.describe(include="all"))
```

```python
numeric_df = df.select_dtypes(include=[np.number])

# Now compute numeric stats on the numeric dataframe
print(numeric_df.skew())
print(numeric_df.kurtosis())
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Option A: Histograms
plt.figure(figsize=(12, 8))
df[numeric_cols].hist(figsize=(12, 8), bins=20)
plt.tight_layout()
plt.show()

# Option B: Density (KDE) Plots per numeric column
# (You might want to do this in a loop or just for selected columns)
for col in numeric_cols:
    plt.figure()
    sns.kdeplot(data=df, x=col, fill=True)
    plt.title(f"Density Plot for {col}")
    plt.show()
```

```python
# Single feature boxplot vs. categories
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, x='category', y=col)  # Replace 'category' with your actual target/label column
    plt.title(f"Boxplot of {col} by Category")
    plt.show()
```

```python
sns.pairplot(df, vars=numeric_cols, hue='category', diag_kind='kde')  # color by your category column
plt.show()
```

```python
subset_cols = ['brightness','saturation','noise_level','blur_metric']
sns.pairplot(df, vars=subset_cols, hue='category', diag_kind='kde')
plt.show()
```

```python
corr_matrix = df[numeric_cols].corr()  # default is Pearson correlation
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
# Select numeric features only
numeric_cols = [
    'brightness', 'saturation', 'noise_level', 'blur_metric',
    'edge_strength', 'motion_blur',
    'lbp_mean_r1', 'lbp_var_r1', 'lbp_mean_r2', 'lbp_var_r2',
    'lbp_mean_r3', 'lbp_var_r3',
    'edges_mean', 'edges_var',
    'red_mean', 'red_var',
    'green_mean', 'green_var',
    'blue_mean', 'blue_var'
]
X = df[numeric_cols].values
y = df['category'].values  # Your target labels

# 2) Scale the data (often recommended before PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3) PCA with ALL possible components (up to #features, or use n_components='mle')
pca_full = PCA()  
pca_full.fit(X_scaled)

# 4) Explained variance ratios per principal component
variance_ratios = pca_full.explained_variance_ratio_
print("Explained variance ratio per component:")
for i, vr in enumerate(variance_ratios, start=1):
    print(f"PC{i}: {vr:.4f}")

# 5) Cumulative explained variance plot
cumulative_var = np.cumsum(variance_ratios)
plt.figure(figsize=(8,5))
plt.plot(range(1, len(cumulative_var)+1), cumulative_var, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()

# 6) If you want a 3D PCA plot to capture “more detail” than 2D:
pca_3d = PCA(n_components=3)
X_pca_3 = pca_3d.fit_transform(X_scaled)
df_pca_3d = pd.DataFrame(X_pca_3, columns=['PC1', 'PC2', 'PC3'])
df_pca_3d['category'] = y

# 7) 3D scatter plot
from mpl_toolkits.mplot3d import Axes3D  # Usually needed for 3D projection

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

# Map your categories to numeric colors. 
# For example, if you have 4 categories:
cat_map = {'clear':0, 'synthetic_haze':1, 'synthetic_low_light':2, 'synthetic_rain':3}
colors = df_pca_3d['category'].map(cat_map)

scatter = ax.scatter(
    df_pca_3d['PC1'], df_pca_3d['PC2'], df_pca_3d['PC3'], 
    c=colors, cmap='viridis', alpha=0.8
)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('3D PCA Visualization')

# Create a legend manually using the category names
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label=k, markerfacecolor='gray', 
           markersize=8) for k in cat_map
]
ax.legend(legend_elements, cat_map.keys(), loc='upper right')
plt.show()
```
